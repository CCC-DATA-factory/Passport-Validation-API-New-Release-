```python
import cv2
from skimage import io
from mrzscanner import MRZScanner

# Create model
model = MRZScanner()

# Read image from the web
img = io.imread('templates\passporty.jpg')
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

# Model inference
result = model(img, do_center_crop=False, do_postprocess=True)

# Output result
print(result)
# {
#     'mrz_polygon':
#         array(
#             [
#                 [ 158.536 , 1916.3734],
#                 [1682.7792, 1976.1683],
#                 [1677.1018, 2120.8926],
#                 [ 152.8586, 2061.0977]
#             ],
#             dtype=float32
#         ),
#     'mrz_texts': [
#         'PCAZEQAQARIN<<FIDAN<<<<<<<<<<<<<<<<<<<<<<<<<',
#         'C946302620AZE6707297F23031072W12IMJ<<<<<<<40'
#     ],
#     'msg': <ErrorCodes.NO_ERROR: 'No error.'>
# }
```

```python
import pytesseract
from passporteye.mrz.text import MRZ
import json


# ✅ Combine MRZ lines into one string (as expected by PyMRZ)
mrz_string = "\n".join(result["mrz_texts"])
print(mrz_string)

# ✅ Parse using PyMRZ
parsed = MRZ(mrz_string)

# ✅ Convert to dict for readability
passport_data = parsed.to_dict()

# Print the parsed fields
for key, value in passport_data.items():
    print(f"{key}: {value}")
```

```python
# Directly from your MRZScanner result
mrz_lines = result["mrz_texts"]  # must be a list of 2 lines

# ✅ Parse directly
mrz = MRZ(mrz_lines)

# Convert to dictionary
passport_data = mrz.to_dict() if mrz else {}

# Display results
print(json.dumps(passport_data, indent=4))
```

```python
from mrzscanner import MRZScanner, ModelType

model = MRZScanner(
   model_type=ModelType.two_stage,
   detection_cfg='20250222',
   recognition_cfg='20250221'
)


img = io.imread('templates\cin.jpg')
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

result = model(img, do_center_crop=False, do_postprocess=False)
print(result)

from capybara import draw_polygon, imwrite, centercrop

poly_img = draw_polygon(img, result['mrz_polygon'], color=(0, 0, 255), thickness=5)
imwrite(centercrop(poly_img))
```

```python
from capybara import imwarp_quadrangle, imwrite
import numpy as np


polygon = np.array([
    [ 158.536 , 1916.3734],
    [1682.7792, 1976.1683],
    [1677.1018, 2120.8926],
    [ 152.8586, 2061.0977]
], dtype=np.float32)

img = io.imread('templates\passport_cote.jpg')
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

mrz_img = imwarp_quadrangle(img, polygon)
imwrite(mrz_img)
```
